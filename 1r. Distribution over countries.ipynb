```r
library(dplyr)
library(ggplot2)
library(readxl)
library(stringr)
```

```r
paintings= read.csv2(file.path('Data', 'PaintingsWithCountryInfo-EN.csv'))

# converting Netherlands to a single location
paintings <- paintings %>%
    mutate(OriginCountry= str_replace(OriginCountry, '^Netherlands.*', 'Netherlands'))

country.count <- paintings %>% 
    dplyr::group_by(OriginCountry) %>%
    dplyr::summarize(count= n()) %>%
    dplyr::arrange(-count)
country.count
```
